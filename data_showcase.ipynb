```python
from datetime import datetime

application['dob'] = pd.to_datetime(application['dob'], errors='coerce')

print(" :", application['dob'].isna().sum())

#  
today = pd.to_datetime("today")

application['age'] = (today - application['dob']).dt.days // 365

application[['dob', 'age']].head()


application['age'].hist(bins=50, figsize=(8,5))
plt.title(" ")
plt.xlabel("")
plt.ylabel("")
plt.show()
```
